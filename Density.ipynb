```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the Seaborn tips dataset
tips = sns.load_dataset("tips")

# Extract the 'tip' column
tip_data = tips["tip"]

# Compute statistics
mean_tip = tip_data.mean()
median_tip = tip_data.median()
quantiles = tip_data.quantile([0.25, 0.75])

# Create the density plot
plt.figure(figsize=(10, 6), facecolor='lightgray',)
kde_plot = sns.kdeplot(tip_data, color='royalblue', fill=True, alpha=0.7)

# Add vertical lines for statistics using vlines with decreasing heights
y_max = plt.gca().get_ylim()[1]
heights = [0.9, 0.8, 0.7, 0.6]  # Decreasing heights for vlines
stats_values = [quantiles[0.25], median_tip, mean_tip, quantiles[0.75]]
colors = ['purple', 'green', 'red', 'orange']
labels = ['25th Percentile', 'Median', 'Mean', '75th Percentile']

for val, color, height, label in zip(stats_values, colors, heights, labels):
    plt.vlines(val, 0, y_max * height, color=color, linestyle='dotted', label=f'{label}: {val:.2f}')
    plt.text(val, y_max * height, f'{val:.2f}', color=color, ha='center', fontsize=10, fontweight='bold')

# Set labels and title
plt.xlabel("Tip Amount ($)", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.title("Density Plot of Tips", fontsize=14, fontweight='bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend()

# Style the axes with visible horizontal lines
#plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add footnote
plt.figtext(0.1, -0.1, "Data Source: Seaborn Tips Dataset", wrap=True, horizontalalignment='left', fontsize=10, color='gray')

# Show the plot
plt.show()
```
