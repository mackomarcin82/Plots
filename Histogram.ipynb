```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the Seaborn tips dataset
tips = sns.load_dataset("tips")

# Extract the 'tip' column
tip_data = tips["tip"]

# Compute statistics
mean_tip = tip_data.mean()
median_tip = tip_data.median()
quantiles = tip_data.quantile([0.25, 0.75])
IQR = quantiles[0.75] - quantiles[0.25]
outlier_threshold = quantiles[0.75] + 1.5 * IQR

# Define bin edges and labels
bins = np.linspace(tip_data.min(), tip_data.max(), 16)
bin_labels = [f"{bins[i]:.2f}" for i in range(len(bins))]

# Create the histogram plot
plt.figure(figsize=(10, 6),facecolor='lightgrey')
kde_plot = sns.histplot(tip_data, bins=bins, kde=True, color='royalblue', edgecolor='black', alpha=0.7)



# Add vertical lines for statistics
plt.axvline(mean_tip, color='red', linestyle='dashed', label=f'Mean: {mean_tip:.2f}')
plt.axvline(median_tip, color='green', linestyle='dashed', label=f'Median: {median_tip:.2f}')
plt.axvline(quantiles[0.25], color='purple', linestyle='dotted', label=f'25th Percentile: {quantiles[0.25]:.2f}')
plt.axvline(quantiles[0.75], color='orange', linestyle='dotted', label=f'75th Percentile: {quantiles[0.75]:.2f}')

# Find KDE estimate for the outlier threshold
kde_x = kde_plot.lines[0].get_xdata()
kde_y = kde_plot.lines[0].get_ydata()
outlier_y = np.interp(outlier_threshold, kde_x, kde_y)


# Add an arrow to mark Q3 + 1.5*IQR
plt.annotate(f'Outlier Threshold: {outlier_threshold:.2f}',
             xy=(outlier_threshold, outlier_y),
             xytext=(outlier_threshold + 1, outlier_y + 0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=10, color='black')

# Set labels and title
plt.xlabel("Tip Amount ($)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Distribution of Tips", fontsize=14, fontweight='bold')
plt.xticks(bins, bin_labels, rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend()

# Style the axes with visible horizontal lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add footnote
plt.figtext(0.1, -0.1, "Data Source: Seaborn Tips Dataset", wrap=True, horizontalalignment='left', fontsize=10, color='gray')


# Show the plot
plt.show()
```
